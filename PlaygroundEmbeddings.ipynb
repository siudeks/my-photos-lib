```python
# Dwa zdania do porównania
sentence1 = "To jest pierwsze zdanie testowe"
sentence2 = "To jest zdanie testowe pierwsze"

from sklearn.metrics.pairwise import cosine_similarity


import ollama
import numpy as np
from scipy.spatial.distance import cosine

# Function to get embeddings from the Ollama server
def get_embeddings(text):
    response = ollama.embeddings(model="llama3.1:8b", prompt=text)
    # display(response)
    return response['embedding']

# Example texts
text1 = "This is the first sentence."
text2 = "This is the second sentence."

# Get embeddings
embedding1 = get_embeddings(text1)
embedding2 = get_embeddings(text2)

print("Embedding for text1:", embedding1)
print("Embedding for text2:", embedding2)

similarity1 = cosine_similarity([embedding1], [embedding2])
print(f"Podobieństwo kosinusowe1: {similarity1[0][0]}")


from sentence_transformers import SentenceTransformer
import numpy as np

# Załaduj model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')


# Wygeneruj osadzenia
embedding1 = model.encode(sentence1)
embedding2 = model.encode(sentence2)
display(embedding1.__class__)
# Porównaj osadzenia za pomocą kosinusowej miary podobieństwa
similarity2 = cosine_similarity([embedding1], [embedding2])

print(f"Podobieństwo kosinusowe2: {similarity2[0][0]}")
```
